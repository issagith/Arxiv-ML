```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('extractions/CSV/BySubCats/20250307_094141.csv')
df
```

```python
print(len(df["category"].unique()))
df.category.unique()
```

```python
df_sorted = df.sort_values(by='category', ascending=True)
df_sorted
```

```python
math_df = df[df['category'].str.startswith('math.')]
last_math_article = math_df.iloc[-1]
last_math_article
```

```python
math_oc_count = len(df[df['category'] == 'math.OC'])
print(f"Number of articles in math.OC category: {math_oc_count}")
```

```python
# Count articles per category
category_counts = df['category'].value_counts()

# Create a bar plot
plt.figure(figsize=(15, 8))
plt.bar(category_counts.index, category_counts.values)
plt.xticks(rotation=90)
plt.xlabel('Categories')
plt.ylabel('Number of Articles')
plt.title('Distribution of Articles by Category')

# Add some spacing at the bottom for category labels
plt.tight_layout()

# Display the plot
plt.show()
```

```python
# Get categories with more than 10000 papers
large_categories = category_counts[category_counts > 10000].index

# Filter the dataframe to keep only papers from these categories
large_categories_df = df[df['category'].isin(large_categories)]

print(f"Number of selected categories: {len(large_categories)}")
print("Selected categories and their counts:")
print(category_counts[category_counts > 10000])
print(f"\nTotal number of papers in the filtered dataset: {len(large_categories_df)}")
large_categories_df
```
